```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_path = "/home/ewachmann/REPOS/Masterarbeit/datasets/QM9/out/c7h10o2/pyscf/scf_from_files_test.h5"
with h5py.File(input_path, "r") as f:
    dataset_name = list(f.keys())[0]  # Get the first dataset key
    ds = f[dataset_name][:]  # Load the dataset into a NumPy array

print(ds.shape)
ds = ds[:5] #! because of not yet full initial testset!
mean_d = np.mean(ds, axis=0)
mean_diff_matrix = np.mean(np.abs(ds - mean_d), axis=0)

fig, ax = plt.subplots(figsize=(8, 4), ncols=2)

ax[0].pcolormesh(mean_d, cmap="bwr")#, vmin=-1.1, vmax=1.1)
ax[0].set_title("Mean Density C7H10O2")
ax[1].pcolormesh(mean_diff_matrix, cmap="bwr")#, vmin=-1.1, vmax=1.1)
ax[1].set_title("Abs. diff mean of Densities C7H10O2")
```
