```python
import sys, pickle
from scipy.special import erf
import numpy as np
sys.path.append('../src/')
sys.path.append("../../scripts/")

import matplotlib.pyplot as plt
from plt_config import use_latex, TU_GREY, TU_RED
use_latex()
```

```python
# Define input range
x = np.linspace(-5, 5, 1000)

# Define common activation functions
step = np.heaviside(x, 1)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0, x)
gelu = 0.5 * x * (1 + erf(x / np.sqrt(2)))

size=20
# Plotting
plt.figure(figsize=(10, 4))
plt.plot(x, step, label="Step")
plt.plot(x, sigmoid, label="Sigmoid")
# plt.plot(x, tanh, label="Tanh")
plt.plot(x, relu, label="ReLU")
plt.plot(x, gelu, label="GeLU")

plt.xlabel("$x$", fontsize=size)
plt.ylim(-.5,2)
plt.xlim(-2,2)
plt.xticks(fontsize=size)
plt.yticks(fontsize=size)
plt.ylabel(r"$\Theta(x)$", fontsize=size)
plt.grid(True)
plt.legend(fontsize=size, loc="upper left")
plt.tight_layout()
plt.savefig("../../fig/background/activation_functions.pdf")
plt.show()
```
